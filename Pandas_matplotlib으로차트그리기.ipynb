```python
# 아래 둘중에 하나 사용
# import matplotlib.pyplot as plt -> FM 선언
# %matplotlib inline -> 잘 인식 안될 경우 위에 선언문 사용

# plt.figure(figsize=(10,6))
# plt.plot(t, y)
# plt.grid()  : 격자 표시해라...
# plt.xlabel('time')
# plt.ylabel('Amplitude')
# plt.title('Sinewave Demo')
# plt.show() : show()라는 함수를 구지 호출할 필요 없다. GUI 버튼 활용 차트 그릴경우 호출해줘야 한다.


import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline

plt.figure(figsize=(10,6)) # 비워있는 그림판을 생성 (행, 열) size
t = np.arange(0, 12, 0.01) # arange : 0~12 사이 step 0.01 간격 실수 생성..
y = np.sin(t)
plt.plot(t, y)  # plot : 라인 그래프 (x, y=sin(x)) 그려라
```

```python
plt.figure(figsize=(10,6))
t = np.arange(0, 12, 0.01)
y = np.sin(t)
plt.plot(t, y)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Sinewave Demo')
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle='--')
l, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2)
l.set_linestyle(':')
```
